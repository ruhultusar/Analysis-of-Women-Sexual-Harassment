# Report on Women Sexual Harrasment (2001-2021)
In this research, we will attempt to examine the state of women's sexual harassment in Bangladesh.
###### Note: Children represent girls under the age of 18 and women represent girls over the age of 18.

```python
# importing different python libraies to analize the data and visualize it properly.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the data set 

stalking_df= pd.read_csv('Sexual Harrasment\Stalking_2011-2021.csv')
attempt_to_rape_df= pd.read_csv('Sexual Harrasment\Attempt-to-Rape_2014-2021.csv')
gang_rape_df= pd.read_csv('Sexual Harrasment\Rape_GangRape_2007-2021.csv')
total_rape_df= pd.read_csv('Sexual Harrasment\Rape_Total_2001-2021.csv')
killed_after_rape_df= pd.read_csv('Sexual Harrasment\Rape_KilledAfterRape_2007-2021.csv')
suicide_after_rape_df= pd.read_csv('Sexual Harrasment\Rape_CommittedSuicideAfterBeingRaped_2007-2021.csv')
```

### Victims of Stalking for last 11 years.(Only Reported Case)

```python
print('Average victims of stalking is {:.2f}'.format(stalking_df.victims.mean()))
```

```python
fig, axes= plt.subplots(1,2, figsize=(15,5))
current_palette = sns.color_palette('colorblind')
current_palette1 = sns.color_palette("RdPu", 10)

axes[0].set_title("Number of stalking victims over the years")
sns.barplot(x="year", y="victims", data= stalking_df, color= current_palette1[5], ax=axes[0])
axes[0].axhline(281,color=current_palette[2],alpha=.6,ls='--')

ax=sns.barplot(x="year", y="protesting_male_attacked", data= stalking_df, color = current_palette[2], alpha=.7,label='Male',ax=axes[1])
ax=sns.barplot(x="year", y="protesting_female_attacked", data= stalking_df, color = current_palette[8], alpha=.9,label='Female',ax=axes[1])
axes[1].set_ylabel("Victims(Attacked after protesting)")
axes[1].legend(frameon = False)
axes[1].set_title("Male VS Frmale (Attacked after protesting)");
```

#### Analysis: 
An average of 281 women have been stalked during the past 11 years. The stalk victim graph demonstrates that while stalking rates peaked in 2011, they subsequently fell over time. The other graph demonstrates that individuals who protest stalking are attacked. Although the tendency of these attacks is also declining, one thing is constant: when people protest against stalking, stalkers prefer to target men more frequently than women.

### Attempt to Rape vs Actual Rape Case

```python
total_rape_14_21_df= total_rape_df[['Years','Total_victims']].drop(total_rape_df.index[8:22],axis=0)
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10, 5))
plt.title("Attempt to Rape vs Actual Rape Case Register")
plt.plot(attempt_to_rape_df.year,attempt_to_rape_df.number,'s--b');
plt.plot(total_rape_14_21_df.Years,total_rape_14_21_df.Total_victims,'o-r');
plt.legend(['Attempt To Rape','Total Rape'])
plt.xlabel('Year')
plt.ylabel('Registered Case');
```

#### Analysis:
This graph showes that  resigtered attempt to rape case is much lower than registered actual Rape case. Assuming that there are  two resons behind it:
- Victims of attemt to rape cases are less likely to file a complain.
- Actually the rape case is much highter than attemp to rape case.

### Trend Analysis of Rape in Bangladesh from 2001 to 2022

```python
ax= sns.lineplot(x='Years', y='Total_victims',data=total_rape_df,color= 'Black');
sns.set(rc = {'figure.figsize':(10,5)})
ax.set_xlim(2000,2023);
```

#### Analysis:
The graph above shows that when the rape rate reaches its peak, the rate drops immediately after that year. The rate peaked in 2022, and we may anticipate a decline in the rate over the ensuing years.

### Ratio of Children & Women (Total Rape VS Gang Rape)

```python
fig, axes= plt.subplots(1,2, figsize=(15,5))

ax= sns.lineplot(x='Years', y='children',data=total_rape_df,color= current_palette[2],ax=axes[0], label= 'Children');
ax= sns.lineplot(x='Years', y='women',data=total_rape_df,color= current_palette[4],ax=axes[0],label= 'Women');
ax= sns.lineplot(x='Years', y='Unidentified_age',data=total_rape_df,color= current_palette[3],ax=axes[0],label= 'Unidentified');
ax.set_xlim(2000,2023);
axes[0].legend(frameon = False)
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Registered Case');
axes[0].set_title("Children Vs Women (Total Rape Case)");

ax= sns.lineplot(x='Years', y='children',data=gang_rape_df,color= current_palette[2],ax=axes[1], label= 'Children');
ax= sns.lineplot(x='Years', y='women',data=gang_rape_df,color= current_palette[4],ax=axes[1],label= 'Women');
ax= sns.lineplot(x='Years', y='Unidentified_age',data=gang_rape_df,color= current_palette[3],ax=axes[1],label= 'Unidentified');
ax.set_xlim(2007,2023);
axes[1].legend(frameon = False)
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Registered Case');
axes[1].set_title("Children Vs Women (Gang Rape Case)");
```

#### Analysis:
From the above graph, it is evident that the rate of women was much higher than children. Suddenly, in 2007, the overall rape rate of children unexpectedly surplus the overall women's rape rate .

On the other hand, the gang rape rate of women is higher than that of children over the years, but the difference between the gang rape rates of women and children is very close.

### Attemp to rape case coampre to actual rape case (2014-2021)

```python
plt.figure(figsize=(10,5));
ax= sns.barplot(x='Years', y='Total_victims',data=total_rape_14_21_df,color = current_palette[7],label='Actual Rape')
ax= sns.barplot(x='year', y='number',data=attempt_to_rape_df,color = current_palette[0],label='Attempt to Rape')
plt.legend(frameon = False)
plt.title('Actual Rape vs Attempt to Rape')
plt.ylabel('Number of victims')
plt.xlabel('Years')
plt.axhline(141,color=current_palette[2],alpha=.9,ls='--');
```

```python
print("Average Rape case filed from 2014 t0 2021 is {:.2f}".format(total_rape_14_21_df.Total_victims.mean()))
print("Average Attempt to Rape case filed from 2014 t0 2021 is {:.2f}".format(attempt_to_rape_df.number.mean()))
```

#### Analysis:
For the last eight years, the average rape case field has been 957, which is significantly higher than the attempted rape cases filed during this time. The graph above demonstrates that there is never a year when the number of attempted rapes is higher than or sufficiently near to the number of actual rapes. There are a few causes for this:

- Most of the time, a rapist succeeds in his attempts to rape a victim.
- People are less likely to report attempted rape instances because they don't take them seriously.
- In Bangladesh, many people hesitate to report rapes out of concern for the reaction of the community. A complaint for an attempted rape is a luxury in this place.
- The number of actual rape attempts is substantially lower.

### Suicide & Killed rate after rape among children and women (2007-2021)

```python
fig, axes= plt.subplots(1,2,figsize=(15,5))
sns.set_style('darkgrid')

ax=sns.lineplot(x='Years', y='women',data= suicide_after_rape_df,ax=axes[0],label= 'Suicide',color= current_palette[2])
ax=sns.lineplot(x='Years', y='women',data= killed_after_rape_df,ax=axes[0],label= 'Killed',color= 'red')
axes[0].legend(frameon = False)
axes[0].set_title("Killed vs Suicide (Women)");

ax=sns.lineplot(x='Years', y='children',data= suicide_after_rape_df,ax=axes[1],label= 'Suicide',color= current_palette[2])
ax=sns.lineplot(x='Years', y='children',data= killed_after_rape_df,ax=axes[1],label= 'Killed',color= 'red')
axes[1].legend(frameon = False)
axes[1].set_title("Killed vs Suicide (Children)");
```

#### Analysis:

- The suicide rate of both female and children is much lower than the murder rate. 

### Suicide Tendedncy of Women & Children after Rape

```python
fig, axes=plt.subplots(1,2,figsize=(16,5))
ax=sns.barplot(x='Years', y='children',data= suicide_after_rape_df,ax=axes[0],label= 'Children',color= current_palette[2], alpha=.7)
ax=sns.barplot(x='Years', y='women',data= suicide_after_rape_df,ax=axes[0],label= 'Women',color= 'red',alpha=.7)
axes[0].set_xlabel("Number of Victims")
axes[0].legend(frameon = False)
axes[0].set_title("Women vs Children(Suicide After Rape)");

ax=sns.scatterplot(x='Years', y='children',data= suicide_after_rape_df,ax=axes[1],label= 'Children',color= current_palette[2])
ax=sns.scatterplot(x='Years', y='women',data= suicide_after_rape_df,ax=axes[1],label= 'Women',color= 'red')
axes[1].set_xlabel("Number of Victims")
axes[1].legend(frameon = False)
axes[1].set_title("Women vs Children(Killed After Rape)");
```

#### Analysis:
- Suicide tendency of children after rape is much higher than the women.
- From the scatter plot it is clear that the suicide data set dosen't follow any pattern.

### Victim killed after rape (women vs children)

```python
fig, axes=plt.subplots(1,2,figsize=(18,5))

ax=sns.barplot(x='Years', y='women',data= killed_after_rape_df,ax=axes[0],label= 'Women',color= 'red')
ax=sns.barplot(x='Years', y='children',data= killed_after_rape_df,ax=axes[0],label= 'Children',color= current_palette[2], alpha=.7)
axes[0].set_xlabel("Number of Victims")
axes[0].legend(frameon = False)
axes[0].set_title("Women vs Children(Killed After Rape)");

ax=sns.scatterplot(x='Years', y='children',data= killed_after_rape_df,ax=axes[1],label= 'Children',color= current_palette[2])
ax=sns.scatterplot(x='Years', y='women',data= killed_after_rape_df,ax=axes[1],label= 'Women',color= 'red')
axes[1].set_xlabel("Number of Victims")
axes[1].legend(frameon = False)
axes[1].set_title("Women vs Children(Killed After Rape)");
```

#### Analysis:
- Rapists used to kill more women after rape than children, but in recent years, more children have been killed than women.
- The scatter plot reveals that, while the killing rate for women is declining, it is more or less stable for children.

### Total Rape Compare to Total Death After Rape

```python
total_rape_df['Total_death']=killed_after_rape_df.Total_victims+ suicide_after_rape_df.Total_victims
total_rape_df_07_21= total_rape_df.drop(total_rape_df.index[15:21],axis=0)
```

```python
plt.subplots(figsize=(15,5))
ax=sns.barplot(x="Years", y="Total_victims", data= total_rape_df_07_21, color = current_palette[2], alpha=.7,label='Total Victims')
ax=sns.barplot(x="Years", y="Total_death", data= total_rape_df_07_21, color = 'red', alpha=.9,label='Total Death After Rape')
plt.ylabel("Victims")
plt.legend(frameon = False)
plt.title("Total Victims VS Total Death");
```

#### Analysis:
The total number of deaths after rape is substantially lower than the total number of rapes, as shown in the graph above. There could be a number of causes:
- Following a suicide, people are less likely to register a complaint (as suicides are also included in total deaths).

- Police were unable to determine the true motive behind the murder after rapping. Because of this, it is listed as a typical murder.
